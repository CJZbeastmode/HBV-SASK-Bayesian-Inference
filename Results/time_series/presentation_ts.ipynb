```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
# TODO: Change root path
root = "/Users/jay/Desktop/Bachelorarbeit"

df_banff = pd.read_csv(
    f"{root}/Implementation/dependencies/hbv_sask/data/Banff_Basin/streamflow.inp",
    delimiter="\t",
    header=None,
)
df_banff.columns = ["date", "value"]
df_banff["date"] = pd.to_datetime(df_banff["date"])
df_banff = df_banff.set_index("date")
df_banff = df_banff.reset_index()
df_banff
```

```python
# TODO: Change root path
root = "/Users/jay/Desktop/Bachelorarbeit"

df_oldman = pd.read_csv(
    f"{root}/Implementation/dependencies/hbv_sask/data/Oldman_Basin/streamflow.inp",
    delimiter="\t",
    header=None,
)
df_oldman.columns = ["date", "value"]
df_oldman["date"] = pd.to_datetime(df_oldman["date"])
df_oldman = df_oldman.set_index("date")
df_oldman = df_oldman.reset_index()
df_oldman
```

```python
# Create subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.1)

# Add first time series
fig.add_trace(go.Scatter(x=df_oldman['date'], y=df_oldman['value'], mode='lines', name='Oldman Basin'), row=1, col=1)

# Add second time series
fig.add_trace(go.Scatter(x=df_banff['date'], y=df_banff['value'], mode='lines', name='Banff Basin'), row=2, col=1)

# Update layout
fig.update_layout(height=600, width=800, title_text="Hydrology Data in Time Series")

# Show plot
fig.show()
```
